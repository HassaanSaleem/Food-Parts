```python
import pandas as pd
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, DirectoryIterator
from skimage.feature import hog
import cv2
```

```python
class ImageWithNames(DirectoryIterator):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.filenames_np = np.array(self.filepaths)

    def _get_batches_of_transformed_samples(self, index_array):
        return (super()._get_batches_of_transformed_samples(index_array),
                self.filenames_np[index_array])
```

```python
train_datagen = ImageDataGenerator()
train_generator = ImageWithNames("F:/FYP/Data/PFDIC-20 O",
                         train_datagen,
                         target_size=(128,128),
                         color_mode='rgb',
                         batch_size=100,
                         class_mode='categorical',
                         shuffle=False)
```

```python
HOG_X={}
HOG_Y={}
imgName={}
count1=0
count2=0
count3=0

for j in range(len(train_generator)):
    xy,file=train_generator[j]
    x,y=xy
    
    for c1 in x:
        
        #temp=hog(cv2.cvtColor(c1,cv2.COLOR_BGR2GRAY).astype("double"))
        temp=hog(c1, orientations=8, pixels_per_cell=(16,16),cells_per_block=(1, 1), multichannel=True)

        HOG_X[count1]=temp.flatten()
        count1+=1
        
    for c2 in y:
        HOG_Y[count2]=c2
        count2+=1
        
    for c3 in file:
        imgName[count3]=c3
        count3+=1
```
