```python
X_train, X_test, y_train, y_test = train_test_split(HighHOG.iloc[:,1:2562], HighHOG['Label_y'],test_size=0.30,random_state=27)
```

```python
clf = svm.SVC(coef0=1.0,random_state=20)
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```
